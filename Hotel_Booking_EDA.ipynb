# Hotel Bookings Dataset - Exploratory Data Analysis (EDA)

## Overview:

The Hotel Bookings dataset provides information about hotel bookings, including various features such as booking details, guest information, and reservation status. The dataset encompasses a wide range of variables, allowing for an in-depth exploration of trends, patterns, and insights related to hotel booking dynamics.

## Dataset Features:

- **Hotel Information:**
  - Types of hotels: city hotels and resort hotels.
  - Arrival details: year, month, week number, and day of the month.

- **Booking Characteristics:**
  - Lead time: number of days between booking and arrival.
  - Number of stays in weekend nights and weeknights.
  - Number of adults, children, and babies.
  - Previous booking history.

- **Reservation Details:**
  - Reserved and assigned room types.
  - Deposit type.
  - Meal type.

- **Customer Information:**
  - Country of origin.
  - Market segment and distribution channel.
  - Repeated guest status.
  - Customer type.

- **Financials:**
  - Average daily rate (ADR).
  - Special requests.
  - Reservation status.

## Purpose of EDA:

The goal of this Exploratory Data Analysis (EDA) is to gain valuable insights and understanding from the Hotel Bookings dataset. Through systematic exploration, we aim to uncover patterns, trends, and relationships within the data. The EDA process involves:

1. **Data Understanding:**
   - Examining the structure of the dataset.
   - Identifying key features and their significance.

2. **Data Cleaning and Wrangling:**
   - Addressing missing values.
   - Handling outliers and anomalies.
   - Optimizing data types for efficient analysis.

3. **Exploratory Analysis:**
   - Conducting univariate analysis to understand the distribution of individual variables.
   - Performing bivariate and multivariate analyses to explore relationships between variables.
   - Visualizing trends and patterns using charts and graphs.

4. **Answering Key Questions:**
   - Investigating hotel preferences, temporal trends, meal choices, and country insights.
   - Analyzing market segments, customer types, lead time impact, and cancellation patterns.
   - Understanding correlations between different features.

5. **Deriving Insights:**
   - Formulating conclusions based on EDA findings.
   - Extracting actionable insights for stakeholders in the hospitality industry.

**Dataset Details**

1. **hotel**: Indicates the type of hotel (e.g., "Resort Hotel" or "City Hotel").

2. **is_canceled**: Binary variable indicating whether the booking was canceled (1) or not (0).

3. **lead_time**: Number of days between booking and arrival.

4. **arrival_date_year**: The year of arrival.

5. **arrival_date_month**: The month of arrival.

6. **arrival_date_week_number**: The week number of arrival.

7. **arrival_date_day_of_month**: The day of the month of arrival.

8. **stays_in_weekend_nights**: Number of weekend nights (Saturday or Sunday) the guest stayed.

9. **stays_in_week_nights**: Number of week nights (Monday to Friday) the guest stayed.

10. **adults**: Number of adults in the booking.

11. **children**: Number of children in the booking.

12. **babies**: Number of babies in the booking.

13. **meal**: Type of meal booked (e.g., "BB" for Bed & Breakfast).

14. **country**: Country of origin of the guest.

15. **market_segment**: Market segment designation (e.g., "Online Travel Agents" or "Direct").

16. **distribution_channel**: Booking distribution channel (e.g., "Online Travel Agents" or "Direct").

17. **is_repeated_guest**: Binary variable indicating whether the guest is a repeated guest (1) or not (0).

18. **previous_cancellations**: Number of previous bookings that were canceled by the guest.

19. **previous_bookings_not_canceled**: Number of previous bookings that were not canceled by the guest.

20. **reserved_room_type**: Type of room reserved by the guest.

21. **assigned_room_type**: Type of room assigned to the guest.

22. **booking_changes**: Number of changes made to the booking.

23. **deposit_type**: Type of deposit made for the booking.

24. **agent**: ID of the travel agency making the booking.

25. **company**: ID of the company or entity making the booking.

26. **days_in_waiting_list**: Number of days the booking was in the waiting list before it was confirmed.

27. **customer_type**: Type of booking customer (e.g., "Transient" or "Contract").

28. **adr**: Average Daily Rate, the average rental income per paid occupied room.

29. **required_car_parking_spaces**: Number of car parking spaces required by the guest.

30. **total_of_special_requests**: Number of special requests made by the guest.

31. **reservation_status**: Reservation last status (e.g., "Canceled", "Check-Out", "No-Show").

32. **reservation_status_date**: Date when the reservation status was last updated.

## Data Wrangling

1. Drop Agent and Company Columns :

    **Reasons**    

    a. **Limited Information Value:**
        - The "company" and "agent" columns serve as identifiers without associated details, limiting their utility for meaningful analysis.

    b. **Significant Null Data Presence:**
        - Both columns have substantial missing values, with the "company" column containing only 5-7% available data, hindering meaningful insights.

    c. **Operational Simplification:**
        - Omitting "company" and "agent" contributes to dataset simplification, reducing complexity and computational overhead for streamlined analysis.

    d. **Focus on Relevant Variables:**
        - Retaining only relevant variables ensures a concise and targeted analysis, emphasizing factors directly contributing to research objectives in hotel bookings.

2. Drop Rows where distribution channels and market segments are undifined (Just 5 such Rows does not impact anything)

3. Drop required car parking space greater than 3 which are 2 rows with just 2 adults and 8 cars parking required

4. 3. Drop lead time greater than 700 which are 2 rows

5. Drop 180 Rows where guests are 0

5. Replace 'Undefined' meals to 'BB' :                        
    **Reasons :**    

    a. **Huge BB contribution in Dataset:**
        - BB contributes to 77.3% of meals column data

    b. **Slight change if added to BB:**
        - Adding Undefined values to BB causes just 1 % change in BB values to 78.29% which is less and expectable

6. Replace Null values in Country to 'PRT' :                        
    **Reasons :**    

    a. **Huge PRT contribution in Dataset:**
        - PRT contributes to 40% of Country column data

    b. **Slight change if added to BB:**
        - Adding PRT values to Null causes just 1% change in PRT values to 41% which is less and expectable

    c. **Customer type wise also correct:**
        - After checking customer type wise PRT comes to be the best choice again

7. Delete Outliers from 'ADR' Columns:

    There are two outliers the first one is having 5400 value 10 times the 2nd highest value and the second is negative which is also not logical dropping both rows is best instead of replacing there values.

8. Fill mean to adr values where 0 is there after outlier deletion :

    a. There are 1959 '0' values in ADR which is just not logical as ADR is Avg Daily Rate which comes after dividing total revenue of day by total rooms in hotel.

    b. The only situation when ADR is 0 is when hotel is empty

    c. Thus filling mean to ADR where it is zero is best option
     

9. Changing few columns datatypes to int16 and few to category to reduce size of data set which was first 28MB+ and now just 9.1MB. This also helps in fast data query.

### Drop Agent and company columns

```python
df.drop(columns = ['agent', 'company'], inplace=True)
df.sample(3)
```

```python
df.shape
```

### Drop Rows where distribution channels and market segments are undifined (Just 5 such Rows does not impact anything)

```python
df.drop(df[(df['market_segment']=='Undefined') | (df['distribution_channel']=='Undefined')].index, inplace=True)
```

```python
df.shape
```

### Drop required car parking above 3

```python
df.drop(df[df['required_car_parking_spaces'] > 3].index, inplace=True)
```

### Drop lead time

```python
df.drop(df[df['lead_time'] > 700].index, inplace=True)
```

### Drop Rows where guests = 0

```python
df.drop(df[(df['adults'] + df['children'] + df['babies']) == 0].index, inplace=True)
```

### Replace 'Undefined' meals to 'BB'

```python
df['meal'] = df['meal'].replace('Undefined', 'BB')
```

```python
df['meal'].unique()
```

### Replace Null values in Country to 'PRT'

```python
df['country'].fillna('PRT', inplace=True)
```

```python
df['country'].isnull().sum()
```

```python
df.isnull().sum()
```

### Delete Outliers from 'ADR' Columns

```python
df.drop(df[(df['adr'] > 1000) | (df['adr'] < 0)].index, inplace = True)
```

```python
df[(df['adr'] > 1000) | (df['adr'] < 0)]
```

### Replace 0 with mean ADR

```python
df['adr'] = df['adr'].replace(0, df['adr'][df['adr'] != 0].mean())
```

```python
df['adr'][df['adr']==0]
```
